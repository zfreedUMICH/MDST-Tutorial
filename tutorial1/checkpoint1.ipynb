Let's compare Michigan to Ohio. Which state has the higher average across all years in the test you chose?

```python
michigan_scores = df[(df['STATE'] == 'MICHIGAN') & (df['GRADES_PK_G'].notna())]['AVG_READING_4_SCORE']
ohio_scores = df[(df['STATE'] == 'OHIO') & (df['GRADES_PK_G'].notna())]['AVG_READING_4_SCORE']
average_michigan_score = michigan_scores.mean()
average_ohio_score = ohio_scores.mean()
if average_michigan_score > average_ohio_score:
    print("Michigan has a higher average across all years in Grade 4 Reading score.")
elif average_michigan_score < average_ohio_score:
    print("Ohio has a higher average across all years in Grade 4 Reading score.")
else:
    print("Michigan and Ohio has the same average across all years in Grade 4 Reading scores.")
```

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
# X =
# y = 
X = df[['instruction_expenditure_per_student', 'AVG_MATH_4_SCORE']].dropna()
y = df.loc[X.index]['AVG_READING_4_SCORE']
y.fillna(y.median(), inplace = True)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(
#      X, y, test_size=, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
# import your sklearn class here
from sklearn.linear_model import LinearRegression
```

```python
# create your model here
# model = 
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(model.score(X_test, y_test))
print(np.mean(model.predict(X_test)-y_test))
print(np.mean(np.abs(model.predict(X_test)-y_test)))
```
